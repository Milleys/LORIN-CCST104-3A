**Exploratory Data Analysis**

**1. Data Loading**

```python
import pandas as pd


df = pd.read_csv('hardwareStore.csv')

print(df.head())

print(df.isnull().sum())

```

**2. Initial Data Analysis**

```python

summary_stats = df.describe()

print(summary_stats)
```

**3. Category Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne
plt.figure(figsize = (12,5))
plt.subplot(1,2,1)
sne.histplot(df['CATEGORY_NAME'],bins=15,kde=True)
plt.title('Distribution of Products Across Categories')


product_number = df.groupby('CATEGORY_NAME')['QUANTITY'].sum()

highest_number_product = product_number.idxmax()

```

**4. Cost and Price Analysis**

```python
average_standard_cost = df['STANDARD_COST'].mean()
average_list_price = df['LIST_PRICE'].mean()





product_standard_cost = df.groupby('PRODUCT_NAME')['STANDARD_COST'].sum()

highest_standard_cost = product_standard_cost.idxmax()




product_highest_price = df.groupby('PRODUCT_NAME')['LIST_PRICE'].sum()

highest_price = product_highest_price.idxmax()

```

**5. Location Analysis**


```python

location = df.groupby('CITY')['QUANTITY'].sum()

highest_location = location.idxmax()

```

**6. Insights and Reporting**




```python
report = """
                           ***Exploratory Data Analysis Report***

                                ***Summary Statistics***\n\n
{}




                              Highest number of Products : {}

                              Average Standard Cost:  {}

                              Average List Price : {}

                              Highest Standard Cost : {}

                              Highest List Price : {}

                              Location with Highest Quantity : {}

""".format(summary_stats, highest_number_product, average_standard_cost, average_list_price, highest_standard_cost, highest_price, highest_location)

print(report)


plt.figure(figsize = (8,6))
sne.barplot(data = df, x = 'CATEGORY_NAME', y = 'LIST_PRICE', color = 'pink')
plt.title('Barplot of CATEGORY NAME AND LIST_PRICE')



plt.figure(figsize = (8,6))
sne.scatterplot(data = df, x = 'LIST_PRICE', y = 'QUANTITY', color = 'purple')
plt.title('Scatterplot of LIST PRICE vs QUANTITY')
```
