Simple Linear Regression

# 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2. Getting Data

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')

```

# 3. Visualizing Data

```python
plt.scatter(data['Previous_Exam_Scores'], data['Grades'])
plt.xlabel('Independent Variable (x)')
plt.ylabel('Dependent Variable (y)')
plt.title('Scatter plot of Data')
plt.show()
```

#4. Preparing for Data Modelling

```python
X = data[['Previous_Exam_Scores']] #independent variable
Y = data[['Grades']] #Dependent variable
```

# 5. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(X,Y)
```

# 6. Getting Model Coefficients

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope : ', slope)
print('Intercept :', intercept)
```

# 7. Visualizing Linear Regression Line

```python
plt.scatter(X,Y, color = 'green')
plt.plot(X,model.predict(X), color = 'red')
plt.xlabel('Grades (x)')
plt.ylabel('Previous Exam Score (y)')
plt.title('Linear Regression')
plt.show()
```

# 8. Making Predictions

```python
new_X = [[50]]

predicted_Y = model.predict(new_X)

print("Predicted Grades:", predicted_Y)
print("Previous Exam Score: ", new_X)
```
