1. Creating Dataset

```python
import pandas as pd

data = {
    'student_id': range(1,21),
    'gender': ['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'test_date' : pd.date_range(start = '2024-02-23', periods = 20),
    'study_hours':[1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,2,2,5],
    'test_score' :[70,76,75,75,73,77,88,78,95,86,73,75,83,85,86,75,64,75,65,74],
    'attendance' :['Present','Present','Present','Absent','Present','Present','Present','Present','Absent','Present','Present','Present','Present','Absent','Present','Present','Present','Present','Absent','Present']

}


df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index= False)

```

2. Data loading and initial examnination

```python
df = pd.read_csv('student_performance.csv')

print(df.head())

print(df.isnull().sum())
```

3. Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

plt.figure(figsize = (12,5))

plt.subplot(1,2,1)

sne.histplot(df['test_score'],bins=15,kde=True)
plt.title('Distribution of Test Score')

plt.subplot(1, 2, 2)
sne.histplot(df['study_hours'],bins=15,kde=True)
plt.title('Distribution of Study Hours')

plt.figure(figsize = (8,6))
sne.scatterplot(data = df, x = 'study_hours', y = 'test_score')
plt.title('Scatterplot of Study Hours vs Test Score')
```

4. Data Visualization

```python
plt.figure(figsize = (6,4))
sne.countplot(data = df, x = 'gender')
plt.title('Gender Distribution')

plt.figure(figsize = (10,6))

sne.lineplot(data =df, x = 'test_date', y = 'test_score')
plt.title('Trend in Test Scores Over Time')

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()

plt.figure(figsize = (8,6))
sne.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Heat Map')
```

```python
report = """
***Exploratory Data Analysis Report***

***Summary Statistics***
{}
""".format(summary_stats)

print(report)
```
